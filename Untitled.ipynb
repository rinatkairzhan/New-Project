```python
data['SSD'] = data['SSD'].str.replace('SSD', '')
data['SSD'] = data["SSD"].str.replace(' ','')
```

```python
#if data['SSD'].str[-2:] == "ТБ":
#    data['SSD'] = int(data['SSD'].str.extract('(\d+)', expand=False))*1024
#elif data['SSD'].str[-2:] == "ГБ":

#   data["SSD"] = int(data['SSD'].str.extract('(\d+)', expand=False))
import numpy as np

data['SSD'] = data['SSD'].replace('нет', np.nan)
```

```python
import re
def convert_ssd(value):
    if pd.isna(value) or value == 'нет':
        return np.nan
    elif 'ТБ' in value:
        num = re.findall(r'\d+', str(value))
        if num:
            return int(num[0]) * 1024
    elif 'ГБ' in value:
        num = re.findall(r'\d+', str(value))
        if num:
            return int(num[0])
    return np.nan

data['SSD'] = data['SSD'].apply(convert_ssd)
```

```python
data['Price'] = data['Price'].str.replace(' ', '')
data['Price'] = data['Price'].str.replace('₸', '')
```
